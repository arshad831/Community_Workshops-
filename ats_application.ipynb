```python
# Import necessary libraries
import google.generativeai as genai  # For accessing Google's generative AI models
import PyPDF2 as pdf  # For PDF file reading and processing
from google.colab import files  # For file upload functionality in Google Colab
import io  # For handling byte streams

# Function to get a response from Google's Gemini AI model
def get_gemini_response(input_prompt):
    model = genai.GenerativeModel('gemini-pro')  # Initialize the Gemini Pro model
    response = model.generate_content(input_prompt)  # Generate content based on the input prompt
    return response.text  # Return the generated text

# Function to extract text from an uploaded PDF file
def input_pdf_text(uploaded_file):
    try:
        reader = pdf.PdfReader(io.BytesIO(uploaded_file))  # Read the PDF from a byte stream
        text = ""
        for page in reader.pages:  # Iterate through each page in the PDF
            text += page.extract_text() or ''  # Extract and concatenate text from each page
        return text
    except Exception as e:
        return f"Error reading PDF: {e}"  # Return error message in case of failure

# Main function to process the resume against the job description
def process_resume(jd, uploaded_file):
    if uploaded_file is None:
        return "Please upload a resume."  # Check for resume upload
    if not jd:
        return "Please provide a job description."  # Check for job description input

    resume_text = input_pdf_text(uploaded_file)  # Extract text from the uploaded resume
    if resume_text:
        # Prepare the input prompt for the generative AI model
        input_prompt = f"""
        Hey Act Like a skilled or very experienced ATS (Application Tracking System)
        with a deep understanding of tech field, software engineering, data science, data analyst, artificial intelligence
        and big data engineer. Your task is to evaluate the resume based on the given job description.
        You must consider the job market is very competitive and you should provide
        the best assistance for improving the resumes. Assign the percentage Matching based
        on JD and the missing keywords with high accuracy.
        resume: {resume_text}
        description: {jd}

        I want the response in different single string in each separate line  having the structure
        {{"JD Match":"%",
        "MissingKeywords:[]",
         Tips to improve resume to get more match}}
        """
        response = get_gemini_response(input_prompt)  # Get AI-generated evaluation
        return response
    else:
        return "Error processing the resume."  # Error message if resume processing fails

def main():
    jd = input("Please paste the job description here: ")  # Get job description as text input

    print("Please upload your resume (PDF):")  # Prompt user to upload resume
    uploaded_file = files.upload()  # Upload file

    uploaded_file_name = next(iter(uploaded_file))  # Assuming only one file is uploaded

    response = process_resume(jd, uploaded_file[uploaded_file_name])  # Process the resume
    print(response)  # Print the AI's evaluation

# Run the main function if this script is executed
if __name__ == "__main__":
    main()
```
